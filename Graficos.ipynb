```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as libreria
import seaborn

libreria.plot( (1,2,3)  ,   (5,3,4) ) # Diagrama de dispersión-> X frente a una Y

```

```python
libreria.bar( (1,2,3)  ,   (5,3,4) ) # GRafico de barras-> X frente a una Y
```

```python
seaborn.distplot( np.random.poisson(size=1000))
```

```python
seaborn.boxplot( np.random.poisson(size=1000))
```

```python
fig,axs=libreria.subplots(1,2,figsize=(14,5))
seaborn.boxplot( np.random.poisson(size=1000), ax=axs[1])
seaborn.distplot( np.random.poisson(size=1000), ax=axs[0])
```

```python
# 1 variable cualitativa:  Barras o sectores
# 1 variable cuantitativa: Histograma o Boxplot
# 2 variables: 2 cualitativas: Barras apiladas
# 2 variables: 1 cualitativa y 1 cuantitativa: Si la cualitativa es dicotómica (2 valores) Piramide poblacional
#                                              Si la cualitativa no es dicotómica Boxplot multiple
# 2 variables cuantitativas:  Dispersión, scatterplot
```

```python
libreria.plot( (1,2) , (10,9) )
libreria.plot( (3,4) , (11,14) )
libreria.plot( (5,6) , (17,19) )
libreria.plot( (7,8) , (18,20) )
```
