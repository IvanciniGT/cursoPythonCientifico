```python
datos['Date']=datos['Date'].apply(lambda texto: pd.Timestamp(texto))
```

```python
datos['Date'].dtype
```

```python
datos_filtrados=datos[datos['PrecipitationSumInches']!=-1000]
mpl.pyplot.plot(datos_filtrados['Date'],datos_filtrados['PrecipitationSumInches'].cumsum())
```

```python
def saca_por_fecha(dataframe,anyo):
    datos_extraidos=dataframe[ dataframe['Date']<=pd.Timestamp(anyo+'-12-31')]
    return datos_extraidos[datos_extraidos['Date']>=pd.Timestamp(anyo+'-01-01') ]

datos_filtrados=datos[datos['PrecipitationSumInches']!=-1000]
datos_2014=saca_por_fecha(datos_filtrados,'2014')
datos_2015=saca_por_fecha(datos_filtrados,'2015')
datos_2016=saca_por_fecha(datos_filtrados,'2016')
datos_2017=saca_por_fecha(datos_filtrados,'2017')
mpl.pyplot.plot(datos_2014['Date'],datos_2014['PrecipitationSumInches'].cumsum())
mpl.pyplot.plot(datos_2015['Date'],datos_2015['PrecipitationSumInches'].cumsum())
mpl.pyplot.plot(datos_2016['Date'],datos_2016['PrecipitationSumInches'].cumsum())
mpl.pyplot.plot(datos_2017['Date'],datos_2017['PrecipitationSumInches'].cumsum())
```
